# Numpy (algebra)

```python
import numpy as np
C = np.array([[1,2],[3,4]])
print(C)
print(C*C)
print(np.dot(C,C))
```

```python
A = np.array([[n+m for n in range(5)] for m in range(5)])
A
```

```python
B = np.array([[[n+m+k for n in range(5)] for m in range(5)] for k in range(5)])
B
```

```python
v = np.array([0,1,0,1,0])
v
```

```python
B*B
```
